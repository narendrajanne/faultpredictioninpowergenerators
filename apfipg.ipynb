# implementing classification algorithm in detect_train dataset

# spliting the detect_train dataset into two variables 

```python
x=df1.drop("Output (S)", axis=1)
```

```python
y=df1["Output (S)"]
```

```python
x
```

```python
y
```

# splitting the data into train and test 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# implementating decision tree classifier 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

# fitting the model 

```python
dtc.fit(x_train, y_train)
```

```python
y_pred=dtc.predict(x_test)
```

# generating accuracy score and confusion matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn import metrics
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
metrics.plot_confusion_matrix(dtc,x_test,y_test)
```

# generating the tree 

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dtc)
plt.show()
```
